```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simular_PAS(p, n, reps=1):
  """
    Genera y devuelve 'num_trays' trayectorias de un Paseo Aleatorio Simple (PAS).

    Cada trayectoria consiste en 'n' pasos.

    Parámetros:
      p (float): Probabilidad de dar un paso de +1. Debe ser un valor en [0, 1].
      n (int): Número de pasos en cada trayectoria.
      reps (int): Número de trayectorias a generar. Por defecto es 1.

    Devuelve:
      numpy.ndarray: Un array 2D de NumPy donde cada fila es una trayectoria del
      PAS. La forma del array es (reps, n).
  """
  # Generamos 'reps' series de 'n' pasos aleatorios, cada paso es +1 o -1
  # la forma del array es (reps, n)
  pasos = np.random.choice([1, -1], size=(reps, n), p=[p, 1-p])
  # Calculamos las trayectorias acumulando los pasos a lo largo de cada fila
  #la forma del array es (reps,n)
  trayectorias = np.cumsum(pasos, axis=1)

  return trayectorias
```

```python
trayectorias = simular_PAS(p=0.5, n=500, reps=10)
plt.figure(figsize=(10, 6))
for i, tray in enumerate(trayectorias):
    plt.plot(tray, label=f'Trayectoria {i+1}', alpha=0.8)

plt.title('10 Trayectorias de Paseo Aleatorio Simple (p=0.5)')
plt.xlabel('n')
plt.ylabel('X_n')
#plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

Ahora vamos a visualizar un PAS junto con la media teórica y un entorno alrededor de la media de longitud +-1 y +- 2 desviaciones típicas teóricas

```python
n=500
p=0.5
trays = simular_PAS(p, n, reps=20)
pasos = np.arange(1, n + 1)

# Cálculos teóricos
media_teorica = pasos * (2 * p - 1)
sigma_teorica = 2 * np.sqrt(pasos * p * (1 - p))

plt.figure(figsize=(12, 7))

# Intervalos de confianza (sombreados)
plt.fill_between(pasos, media_teorica - 2 * sigma_teorica, media_teorica + 2 * sigma_teorica,
                 color='gray', alpha=0.2, label='±2 Desviaciones típicas')
plt.fill_between(pasos, media_teorica - 1 * sigma_teorica, media_teorica + 1 * sigma_teorica,
                 color='gray', alpha=0.4, label='±1 Desviación típica')

# Media
plt.plot(pasos, media_teorica, color='red', linewidth=1.5, linestyle='--', label='Media teórica', zorder=10)

# Graficamos cada trayectoria
for i, t in enumerate(trays):
    plt.plot(pasos, t, alpha=0.7, linewidth=1)

plt.title(f'20 Trayectorias de PAS (p={p})')
plt.xlabel('n')
plt.ylabel('X_n')
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)
plt.show()
```

Ahora vamos a visualizar la media teórica junto con la media empírica cuando promediamos varias trayectorias.

```python
#calculamos 10000 trayectorias que después promediaremos
p=0.5
trays = simular_PAS(p, n, reps=10000)
num_trays=[5,100,200,500,1000,10000]

#calculo de la media teórica
n=500
pasos=np.arange(1,n+1)
media_teorica=pasos*(2*p-1)

#medias de las trayectorias
medias=[]
for i,num in enumerate(num_trays):
  trayectorias=trays[:num,]
  medias.append(np.mean(trayectorias, axis=0))


#mostrar trayectorias medias
plt.figure(figsize=(12,7))
plt.plot(pasos, media_teorica, color='red', linewidth=1.5, label='Exact', zorder=10)
for i, t in enumerate(medias):
    plt.plot(pasos, t, alpha=0.7, linewidth=1, label=num_trays[i])
plt.xlabel('n')
plt.ylabel('X_n')
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)
plt.show()
```
